```python
from transformers import pipeline
generator = pipeline(
    "text-generation",
    model="gpt2"
)
prompt = "artifical intelligence in finace will"
result = generator(prompt, max_length=50, num_return_sequences=1, truncation=True, max_new_tokens=None)
print("AI Generator text:")
print(result[0]['generated_text'])
```
